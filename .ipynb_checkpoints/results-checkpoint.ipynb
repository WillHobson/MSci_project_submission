```python
#note in this section we are dealing in Torr, and a growth rate of um/min
def Pdcalc(T):
    term1 = (1.27e30)/133
    term2 = T**(-5.65)
    term3 = np.exp(-14640/T)
    return term1*term2*term3

def Keqm(kf, Km, Pd, n):
    rootPd = Pd**(1/2)
    term1 = Km*rootPd
    term2 = 1+term1
    Keq = kf*((term1/term2)**n)
    return Keq

def growthrate(P, Pd, Keq, Km):
    rootPd = Pd**(1/2)
    rootP = P**(1/2)
    term2 = (P/Pd)**(2/2)
    term3 = ((1+(Km*rootPd))/(1+(Km*rootP)))**2
    Ug = Keq*((term2 * term3) - (rootPd/rootP))
    return Ug

def simming(kf, Km, T, n, P):
    Pd = Pdcalc(T)
    #print(f'Pd for T={T} is {Pd}')
    Keq = Keqm(kf, Km, Pd, n)
    Ug = growthrate(P, Pd, Keq, Km)
    return Ug
```

```python
pressures = np.linspace(1, 60000, 6001)

myfit25 = simming(0.223, 0.1653, 298, 2, pressures)
myfit90 = simming(2.61, 0.081, 363, 2, pressures)
myfit200 = simming(27, 0.05,473, 2, pressures)
```

```python
def simulation_const_p_T(P_init, T, M_U, V, Rad_init, t_step, qf, A, shape, df):
    times = np.array(df['Time [ms]'])
    temps = np.array(df['Temp 1 [Â°C]'])+273
    T = temps[0]
    R = 8.31                   #Ideal gas constant
    M_U = 0.9                  #Mass uranium
    mol_U = M_U/238            #Moles Uranium
    rho_U = 19.1e3             #Density Uranium
    imHsys = P_init*V/(R*T)    #initial moles hydrogen in system
    Hmsys = imHsys             #moles of H in system

    if shape == 'sphere':
        geometric_factor = 3
    elif shape == 'cylinder':
        geometric_factor = 4
    else:
        geometric_factor = 3
    SA_init = geometric_factor*((M_U/1000)/rho_U)/Rad_init    #surface area of particle


    MTmolG = 0.0125*M_U/2      #max theoretical absorbed H mols
    MTmolG  = 1.5*mol_U       #max theoretical absorbed H mols


    SA_array = [SA_init]       #array to track surface area
    t_array = [0]              #array to track time
    mol_array = [0]            #array to track mols absorbed
    tot_mol_abs = 0            #variable to track progression
    radii_array = [Rad_init]
    r_hyd = [0]
    T_array = [temps[0]]

    wgs = [bm.bm_ck(T,P_init,qf, A, r_hyd[-1])]  #array to track weight gain rates
    
    

    #while reaction is still ongoing
    while tot_mol_abs/MTmolG < 0.999999:
        #find time value closest to simtime
        ind = np.abs(times - t_array[-1]).argmin()
        T = temps[ind]
        T_array.append(T)
        ##############################################
        P = P_init                           #take const pressure
        wg = bm.bm_ck(T, P,qf, A, r_hyd[-1])           #calculate the weight gain at that pressure
        mol_abs = (wg[0]*SA_array[-1]*t_step)#moles absorbed in one time step
        tot_mol_abs += mol_abs               #add moles absorbed to total mols absorbed
        mol_array.append(tot_mol_abs/MTmolG)
        Hmsys = Hmsys - mol_abs              #calculate mols H left in system
        SA_current = SA_init*(1-(tot_mol_abs/MTmolG))    #change in SA
        r_current = Rad_init*(1-(tot_mol_abs/MTmolG))
        radii_array.append(r_current)
        t_array.append(t_array[-1]+t_step)   #record the time it was here
        wgs.append(wg)                       #record the weight gain rate
        SA_array.append(SA_current)          #record the SA
        if len(mol_array)>100000:
            tot_mol_abs = MTmolG
        r_hyd.append(Rad_init-r_current)
        
    #r_hyd = Rad_init - np.array(radii_array)    
    return t_array, mol_array, radii_array, SA_array, T_array, wgs

jac_new = jac[['Time [ms]', 'Actual flow [L/min]', 'Temp 1 [Â°C]', 'Total flow [L]']]
jac_new['Time [ms]'] = jac_new['Time [ms]']/1000
jac_new['Time [ms]'] = jac_new['Time [ms]']-40
jac_new = jac_new[jac_new['Time [ms]']>=0]

tr = simulation_const_p_T(50000, 479, 160, 9e-5, 50e-6, 1, 41840, 0, 'sphere', jac_new)
trnt = bm.simulation_const_p(50000, 479, 160, 9e-5, 50e-6, 1, 41840, 0, 'sphere')

fig,ax=plt.subplots(figsize=[10,8], dpi=300)
ax.plot(jac_new['Time [ms]'], jac_new['Temp 1 [Â°C]'], linewidth=3, c='red')
al = ax.twinx()
al.plot(jac_new['Time [ms]'], jac_new['Total flow [L]'], linewidth=3, c='tab:green')
al.axhline(y=6.4, color='black', linestyle='--')
ax.tick_params(axis='both', which='both', labelsize=18)
al.tick_params(axis='both', which='both', labelsize=18)
ax.set_xlabel('Time, s', fontsize=18)
ax.set_ylabel('Temperature, K', fontsize=18, c='red')
al.set_ylabel('Actual flow [L]', fontsize=18, c='green')
al.spines['left'].set_color('red')
al.spines['right'].set_color('green')


for tick_label in al.get_yticklabels():
    tick_label.set_color('green')
    

for tick_label in ax.get_yticklabels():
    tick_label.set_color('red')
```
