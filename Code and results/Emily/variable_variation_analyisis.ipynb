## Defining Functions to Implement Models

```python
from models import Stakebake as stb
stb_inst = stb()


def round_decimals_down(number:float, decimals:int=2):
    """
    Returns a value rounded down to a specific number of decimal places.
    """
    if not isinstance(decimals, int):
        raise TypeError("decimal places must be an integer")
    elif decimals < 0:
        raise ValueError("decimal places has to be 0 or more")
    elif decimals == 0:
        return math.floor(number)

    factor = 10 ** decimals
    return math.floor(number * factor) / factor

#performing Stakbake model
def stakebake(K, wm, n):
    """
    stakebake is an implementation of the equation given J. L. Stakebake's paper ‘Kinetics for the reaction of hydrogen with uranium powder’
    INPUT: reaction rate constant, K (/s)
           maximum amount of hydrogen reacted, wm (mg/g)
           reaction order, n (0<n<1)
    OUTPUT: the amount of hydrogen reacted at time t, wt (mg/g)
            time, t (s)
    """
    if n != 1:
        Mo = 1 # in mg/g
        M = Mo**(n-1)

        #calclaute upper limit of t based on stakebake paper in min
        upper_t_lim = (Mo**(1-n))/(K*(1-n))
        upper_t_lim = round_decimals_down(upper_t_lim, 2) # round to 2 dp
                
        #create array of times between 0 and upper lim in steps of 0.1
        t=np.linspace(0,int(upper_t_lim),int(upper_t_lim*100000+1))
        wt = (1- (1+(K*(n-1)*M*t))**(1/(1-n)) )*wm # calculate wm
    else:
        t = np.linspace(0,15, 150)
        wt = wm*(1- np.exp(-K*t))
    return wt, t

import blochmintz as bmm

#performing BlochMintz model
class BlochMintz:
        """
    An implementation of the interfacil growth model equation given J. Bloch and M. H. Mintz's paper ‘Kinetics and mechanism of the U-H reaction’
        INPUT: 
            particle radius, r (micro m)
            temperature, t (K)
            pressure, p (pascals)
            mass of uranium, m (g)
            volume of gas rig, v (m^3)
            constant pressure, const (True or False)
            activation energy, qf (J/mol)
            change in activation energy, A (J/mol)
        OUTPUT: 
            array of times, r[0] (s)
            array of weight gain at time t, r[1] (mg)

    """
    def simulate(self,r,t,p, m, v, const, qf=29300, A=0):
        p = int(p)
        t = int(t)
        r = float(r)/1000000

        if const == True:
            r = bmm.simulation_const_p(p, t, m, v,r, 0.01, qf, A)
        else:
            r = bmm.simulation(p, t, m, v,r, 0.01, qf, A)
            
        return r[0], r[1]

inst_bm_2 = BlochMintz()

    
#Functions calculating the doeff of determ for each dataset respectively    
def coeff_determ_700(model_t, model_y):
    interpolated_gar_y = np.interp(model_t, g700["x"], g700[" y"])
    r2 = r2_score(interpolated_gar_y, model_y)
    return r2

def coeff_determ_550(model_t, model_y):
    interpolated_gar_y = np.interp(model_t, scaled_gar_x_500, flip_norm_gar_y_500)
    r2 = r2_score(interpolated_gar_y, model_y)
    return r2

##optimisation function
def fitandplot(xdat, ydat,ax, params=None):
    if params==None:
        ko, no, r2 = stb.optimize_stakebake(xdat, ydat)
        x,t = stb.stakebake(ko,no)
#         ax.plot(t,x/12.5, label = f'stakebake model, r2={round(r2,3)}')
#         ax.plot(xdat, ydat, label='data')
#         ax.legend(loc='lower right')
#         ax.set_xlim(0, xdat.max())
#         plt.show
    else:
        k,n=params
        ko, no, r2 = stb.optimize_stakebake(xdat, ydat, k,n)
        x,t = stb.stakebake(ko,no)
#         ax.plot(t,x/12.5, label = f'stakebake model, r2={round(r2,3)}')
#         ax.plot(xdat, ydat, label='data')
#         ax.legend(loc='lower right')
#         ax.set_xlim(0, xdat.max())
#         plt.show
        
    return r2, t, x
```
