```python
import numpy as np

# Create an example array with NaN values
arr = np.array([1, np.nan, 3, 4, 4, 6])

# Check if any NaN values exist in the array
nan_indices = np.isnan(arr)

# Find the indices of NaN values
nan_indices_array = np.where(nan_indices)[0]

if len(nan_indices_array) > 0:
    print("NaN values exist in the array.")
    print("Indices of NaN values:", nan_indices_array)
else:
    print("No NaN values in the array.")
```
